# ROOT DIR 설정

```python
ROOT_DIR = '/content'
IMG_EXTENSION = '.jpg'
LABEL_EXTENSION = '.txt'
```

# Clothes Data 압축 해제

```python
import zipfile
import os

%rm -rf {DATA_ROOT_DIR}
%cd {ROOT_DIR}
with zipfile.ZipFile(os.path.join(ROOT_DIR,'Clothes_Data.zip')) as target_file:
    target_file.extractall(os.path.join(ROOT_DIR, 'Clothes_Data'))
```

# Train Test Valid Split

## DATA_ROOT_DIR 설정

```python
DATA_ROOT_DIR = os.path.join(ROOT_DIR, 'Clothes_Data')

TRAIN_ROOT_DIR = os.path.join(DATA_ROOT_DIR,'train')
TEST_ROOT_DIR = os.path.join(DATA_ROOT_DIR,'test')
VALID_ROOT_DIR = os.path.join(DATA_ROOT_DIR,'valid')

TRAIN_IMAGE_DATA_DIR = os.path.join(TRAIN_ROOT_DIR,'images')
TRAIN_LABEL_DATA_DIR = os.path.join(TRAIN_ROOT_DIR,'labels')

TEST_IMAGE_DATA_DIR = os.path.join(TEST_ROOT_DIR,'images')
TEST_LABEL_DATA_DIR = os.path.join(TEST_ROOT_DIR,'labels')

VALID_IMAGE_DATA_DIR = os.path.join(VALID_ROOT_DIR,'images')
VALID_LABEL_DATA_DIR = os.path.join(VALID_ROOT_DIR,'labels')
```

## TRAIN으로 모든 데이터 COPY

```python
import shutil

shutil.copytree(DATA_ROOT_DIR,
                os.path.join(TRAIN_ROOT_DIR))
```

```python
%mkdir {TEST_ROOT_DIR}
%mkdir {os.path.join(TEST_ROOT_DIR,'images')}
%mkdir {os.path.join(TEST_ROOT_DIR,'labels')}

%mkdir {VALID_ROOT_DIR}
%mkdir {os.path.join(VALID_ROOT_DIR,'images')}
%mkdir {os.path.join(VALID_ROOT_DIR,'labels')}
```

## Valid data split

```python
import random

split_ratio = 0.1

img_file_name_list = list(map(lambda x : x.split('.')[0], os.listdir(TRAIN_IMAGE_DATA_DIR)))

split_num = int(len(img_file_name_list) * split_ratio)
random.shuffle(img_file_name_list)

move_file_name_list = img_file_name_list[:split_num]

print('=======================================')
print('train -> valid 10% / start moving...')

for file_name in move_file_name_list:

    shutil.move(os.path.join(TRAIN_IMAGE_DATA_DIR,file_name) + IMG_EXTENSION,
                os.path.join(VALID_IMAGE_DATA_DIR,file_name) + IMG_EXTENSION)
    
    shutil.move(os.path.join(TRAIN_LABEL_DATA_DIR,file_name) + LABEL_EXTENSION,
                os.path.join(VALID_LABEL_DATA_DIR,file_name) + LABEL_EXTENSION)

print('=======================================')
print('train nums : ', len(os.listdir(TRAIN_IMAGE_DATA_DIR)),'\ntrain nums : ', len(os.listdir(TRAIN_LABEL_DATA_DIR)))
print('valid nums : ', len(os.listdir(VALID_IMAGE_DATA_DIR)),'\nvalid nums : ', len(os.listdir(VALID_LABEL_DATA_DIR)))
print('=======================================')
```

## Test data split

```python
import random

split_ratio = 0.1

img_file_name_list = list(map(lambda x : x.split('.')[0], os.listdir(TRAIN_IMAGE_DATA_DIR)))

split_num = int(len(img_file_name_list) * split_ratio)
random.shuffle(img_file_name_list)

move_file_name_list = img_file_name_list[:split_num]

print('=======================================')
print('train -> test 10% / start moving...')

for file_name in move_file_name_list:

    shutil.move(os.path.join(TRAIN_IMAGE_DATA_DIR,file_name) + IMG_EXTENSION,
                os.path.join(TEST_IMAGE_DATA_DIR,file_name) + IMG_EXTENSION)
    
    shutil.move(os.path.join(TRAIN_LABEL_DATA_DIR,file_name) + LABEL_EXTENSION,
                os.path.join(TEST_LABEL_DATA_DIR,file_name) + LABEL_EXTENSION)

print('=======================================')
print('train nums : ', len(os.listdir(TRAIN_IMAGE_DATA_DIR)),'\ntrain nums : ', len(os.listdir(TRAIN_LABEL_DATA_DIR)))
print('test nums : ', len(os.listdir(TEST_IMAGE_DATA_DIR)),'\ntest nums : ', len(os.listdir(TEST_LABEL_DATA_DIR)))
print('=======================================')
```

## TRAIN TEST VALID 개수 확인

```python
print('=======================================')
print('train nums : ', len(os.listdir(TRAIN_IMAGE_DATA_DIR)),'\ntrain nums : ', len(os.listdir(TRAIN_LABEL_DATA_DIR)))
print('\nvalid nums : ', len(os.listdir(VALID_IMAGE_DATA_DIR)),'\nvalid nums : ', len(os.listdir(VALID_LABEL_DATA_DIR)))
print('\ntest nums : ', len(os.listdir(TEST_IMAGE_DATA_DIR)),'\ntest nums : ', len(os.listdir(TEST_LABEL_DATA_DIR)))
print('=======================================')
```
