Sequential API

```python
import tensorflow as tf

from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Flatten, Dense, Input, concatenate
from tensorflow.keras.optimizers import SGD, Adam

import numpy as np
```

```python
model = Sequential()

model.add(Dense(8, input_shape=(4, ), activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))

model.summary()
```

Funcational API

```python
input_1 = Input(shape=(4, ))
hidden_1 = Dense(8, activation='relu')(input_1)
hidden_2 = Dense(16, activation='relu')(hidden_1)
hidden_3 = Dense(32, activation='relu')(hidden_2)

output_ = Dense(10, activation='softmax')(hidden_3)

model = Model(inputs=input_1,outputs=output_)
model.summary()
```

다중 입력 Funcational API

```python
# 입력층 1에 대한 신경망
input_1 = Input(shape=(4,),name='input1')
hidden_1 = Dense(8, activation='relu', name='hidden1')(input_1)
hidden_2 = Dense(16, activation='relu', name='hidden2')(hidden_1)

output_1 = Model(inputs=input_1, outputs=hidden_2)

# 입력층 2에 대한 신경망
input_2 = Input(shape=(8,), name='input2')
hidden_3 = Dense(8, activation='relu', name='hidden3')(input_2)

output_2 = Model(inputs=input_2, outputs=hidden_3)

# Model
result = concatenate([output_1.output, output_2.output], name='concat')

output_ = Dense(10, activation='softmax', name='output')(result)

model = Model(inputs=[input_1, input_2], outputs=output_)
model.summary()
```
