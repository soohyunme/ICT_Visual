# Module import

```python
import numpy as np

def sigmoid(x):
    return 1 / (1+np.exp(-x))
```

# Data Definition 

```python
input = [[2, 1, 2], [1, 2, 3], [3, 0, 2]]
label = [1, 1, 3]
```

```python
input_units = 3
hidden_units = 4
output_units = 5
```

## one hot enccoding

```python
xdata = np.array(input)
tdata = np.zeros([3,5])

for index in range(len(label)):
    tdata[index, label[index]] = 1.0  # int() 해주어서 열 인덱스에 대해 정수형 취한다

print(xdata)

for index in range(len(label)):
    
    print('label = ', label[index], ', one-hot encoding = ', tdata[index])
```

# Initialize weights and bias

```python
W2 = np.random.rand(input_units, hidden_units)
b2 = np.random.rand(hidden_units)

W3 = np.random.rand(hidden_units, output_units)
b3 = np.random.rand(output_units)
```

```python
print(f'xdata.shape = {xdata.shape}, tdata.shape = {tdata.shape}')
print(f'W2.shape = {W2.shape}, b2.shape = {b2.shape}')
print(f'W3.shape = {W3.shape}, b3.shape = {b3.shape}')
```

# Define loss function and output, y

```python
A1 = xdata
A2 = sigmoid(np.dot(A1,W2)+b2)

print(A2.shape)

Y = A3 = sigmoid(np.dot(A2,W3)+b3)

print(A3.shape)
```

# Predict

```python
test_xdata = np.array([1, 1, 2])
```

```python
def predict(test_data):

    A1 = test_data
    A2 = sigmoid(np.dot(A1, W2)+ b2)
    sigmoid_val = sigmoid(np.dot(A2, W3)+ b3)
    Y = np.argmax(sigmoid_val)

    return Y, sigmoid_val # 사람들이 원하는 결과값과 loss를 계산하기 위한 결과 값이 다르므로 확률값과 실제값을 분리
```

```python
predicted_num, sigmoid_val = predict(test_xdata)

print('prediction = ',predicted_num)
print('sigmoid val = ')
print(sigmoid_val.reshape(5,1))
```
