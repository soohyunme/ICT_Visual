```python
import tensorflow as tf

import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
gen = ImageDataGenerator(rescale=1./255)
```

```python
import zipfile

with zipfile.ZipFile('/content/test_class_mode.zip') as target_file:

    target_file.extractall('test_class_mode/')
```

# Binary class mode

```python
data_gen = gen.flow_from_directory(directory='./test_class_mode',
                                   batch_size=3,
                                   shuffle=True,
                                   target_size=(100,100),
                                   class_mode='binary')
print(data_gen.class_indices)
print(data_gen.num_classes)
print(data_gen.class_mode)
```

```python
class_names = {0:'cat',1:'dog'}

img, label = data_gen.next()

for i in range(len(label)):
    print('label => ', label[i])
    
plt.figure(figsize=(8,8))
for i in range(len(img)):
    plt.subplot(1,len(label),i+1)
    plt.xticks()
    plt.yticks()
    animal_name = class_names[label[i]]
    plt.title(animal_name)
    plt.imshow(img[i])
```

# Categorical class mode

```python
data_gen = gen.flow_from_directory(directory='./test_class_mode',
                                   batch_size=3,
                                   shuffle=True,
                                   target_size=(100,100),
                                   class_mode='categorical')
print(data_gen.class_indices)
print(data_gen.num_classes)
print(data_gen.class_mode)
```

```python
class_names = {0:'cat', 1:'dog'}

img, label = data_gen.next()

for i in range(len(label)):
    print('label => ',label[i])

plt.figure(figsize=(8,8))
for i in range(len(img)):
    plt.subplot(1,len(img),i+1)
    plt.xticks()
    plt.yticks()
    animal_name = class_names[np.argmax(label[i])]
    plt.title(animal_name)
    plt.imshow(img[i])
plt.show()
```

# Sparse class mode

```python
data_gen = gen.flow_from_directory(directory='./test_class_mode',
                                   batch_size=3,
                                   shuffle=True,
                                   target_size=(100,100),
                                   class_mode='sparse')

print(data_gen.class_indices)
print(data_gen.num_classes)
print(data_gen.class_mode)
```

```python
class_names = {0:'cat', 1: 'dog'}

img, label = data_gen.next()

for i in range(len(label)):
    print('label => ', label[i])

plt.figure(figsize=(8,8))
for i in range(len(img)):
    plt.subplot(1,len(img),i+1)
    plt.xticks()
    plt.yticks()
    animal_name = class_names[label[i]]
    plt.title(animal_name)
    plt.imshow(img[i])
plt.show()
```
